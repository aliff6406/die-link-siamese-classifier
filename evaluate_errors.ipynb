```python
import pandas as pd

# Load the error analysis data
error_analysis_df = pd.read_csv('./eval_results/error_analysis.csv')

# Concatenate Coin1 and Coin2 columns for overall coin occurrence count
all_coins = pd.concat([error_analysis_df['Coin1'], error_analysis_df['Coin2']], axis=0)

# Count occurrences of each coin
coin_counts = all_coins.value_counts().reset_index()
coin_counts.columns = ['Coin', 'Occurrences']

pd.set_option('display.max_rows', None)
# Print the number of instances for each unique coin
print(coin_counts)
```
